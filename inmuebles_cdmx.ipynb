## **Demandas de ML**

### **¿Cuáles son los valores promedio de alquiler por tipo de propiedad?**

```python
datos.head()
```

```python
datos['Valor'].mean() ##calcula promedio
```

```python
datos.groupby('Tipo')['Valor'].mean()
```

```python
datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

```python
df_tipo_precio = datos.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_tipo_precio.plot(kind='barh', figsize=(12,8), color= 'purple')
```

```python
df_tipo_precio.plot.bar(figsize=(12,8), color= 'purple')
```

```python
datos.Tipo.unique()
```

```python
datos['Tipo'].unique()
```

```python
inmuebles_comerciales = ['Conjunto Comercial/Sala',
       'Tienda/Salón', 'Casa Comercial','Terreno Estándar', 'Cochera/Estacionamiento','Galpón/Depósito/Almacén', 'Tienda en Centro Comercial', 'Hotel', 'Loteo/Condominio',
       'Industria']

datos.query('@inmuebles_comerciales in Tipo')
```

```python
datos.query('@inmuebles_comerciales not in Tipo')
```

```python
df = datos.query('@inmuebles_comerciales not in Tipo')
df.head()
```

```python
df_tipo_precio = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')

df_tipo_precio.plot(kind='barh', figsize=(12,8), color= 'purple')
```

## **Calcular el promedio de habitaciones por departamento.**

```python
df2_promedio_habitaciones = df2.groupby('Tipo')['Habitaciones'].mean()
df2_promedio_habitaciones.name = 'Promedio Habitaciones'
df2_promedio_habitaciones
```

## **Analizar qué colonias tienen el promedio de alquiler más alto.**


```python
df2_colonias_valor = df2.groupby('Colonia')[['Valor']].mean().sort_values('Valor', ascending=False)
df2_colonias_valor.name = 'Promedio Valor'
df2_colonias_valor
```
